```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def my_function(a, d):
    return a - (a - d * -0.5) / (1 + 0.5**2)

# Generate values for a and d
a_values = np.linspace(0, 1, 100)
d_values = np.linspace(0, 1, 100)

# Create a grid of values for a and d
a, d = np.meshgrid(a_values, d_values)

# Calculate the function values for each combination of a and d
z = my_function(a, d)

# Create a contour plot
plt.figure(figsize=(8, 6))
contour_plot = plt.contourf(a, d, z, cmap='viridis')
plt.colorbar(contour_plot, label='Function Value')
plt.xlabel('a')
plt.ylabel('d')
plt.title('Contour Plot of the Function')

# Show the plot
plt.show()
```
